```python
import re
import folium
import webbrowser
import os

# -------- FUNCION PARA EXTRAER COORDENADAS --------
def extraer_coordenadas(link):
    match = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', link)
    if match:
        return float(match.group(1)), float(match.group(2))

    match = re.search(r'q=(-?\d+\.\d+),(-?\d+\.\d+)', link)
    if match:
        return float(match.group(1)), float(match.group(2))

    match = re.search(r'!3d(-?\d+\.\d+)!4d(-?\d+\.\d+)', link)
    if match:
        return float(match.group(1)), float(match.group(2))

    return None


# -------- LISTA DE LINKS --------
links = [
    "https://www.google.com/maps?q=-25.49758529663086,-54.658931732177734&z=17&hl=en",
    "https://www.google.com/maps?q=-25.48276710510254,-54.65186309814453&z=17&hl=en",
    "https://www.google.com/maps/place/25°28'23.6%22S+54°39'00.8%22W/@-25.473291,-54.6502888,21z/data=!4m5!3m4!4b1!8m2!3d-25.4732265!4d-54.6502151",
    "https://www.google.com/maps/place//@-25.4634002,-54.6379959",
    "https://www.google.com/maps/place/25°29'36.9%22S+54°41'05.8%22W/@-25.4936191,-54.6850251",
    "https://www.google.com/maps?q=-25.4896790,-54.6673480",
    "https://www.google.com/maps/place/Mersh+Corporation/@-25.4958806,-54.6620822,21z"
]

# -------- NOMBRES --------
nombres = [
    "Eduardo",
    "Fortín",
    "Subestación",
    "Planta",
    "nodo extra",
    "aaa",
    "KM7"
]

colors = [
    'blue',
    'red',
    'green',
    'orange',
    'purple',
    'darkblue',
    'lightgreen',
    'black'
]


# -------- EXTRAER COORDENADAS --------
nodos = []

for i, link in enumerate(links):
    coord = extraer_coordenadas(link)

    if coord:
        nodos.append({
            "nombre": nombres[i] if i < len(nombres) else f"Punto {i + 1}",
            "lat": coord[0],
            "lon": coord[1]
        })


# -------- ZONA PEQUEÑA -------

zona_eduardo_1 = [
    (-25.504493, -54.648241),
    (-25.514757, -54.652714),
    (-25.508168, -54.670148),
    (-25.508771, -54.668567),
    (-25.514299, -54.670942),
    (-25.513734, -54.672598),
    (-25.497741, -54.667535),
]

# -------- ZONA GRANDE AJUSTADA AL SCREENSHOT VERDE --------
zona_eduardo_2 = [
    (-25.458367, -54.652666),
    (-25.459766, -54.647773),
    (-25.463158, -54.648977),
    (-25.465664, -54.642558),
    (-25.463330, -54.639117),
    (-25.470844, -54.636036),
    (-25.499327, -54.660415),
    (-25.494953, -54.691403),
    (-25.475846, -54.678640),
    (-25.471258, -54.657319),
]

puntos = [
    (-25.486853, -54.635977),
    (-25.495864, -54.639016),
    (-25.487978, -54.662585),
    (-25.479928, -54.656316),
]

# -------- CREAR MAPA --------
if nodos:
    mapa = folium.Map(
        location=[-25.4895, -54.6600],
        zoom_start=14,
        tiles="OpenStreetMap"
    )

    # -------- MARCADORES --------
    for i, nodo in enumerate(nodos):
        folium.Marker(
            location=[nodo["lat"], nodo["lon"]],
            popup=nodo["nombre"],
            tooltip=nodo["nombre"],
            icon=folium.Icon(
                color=colors[i % len(colors)],
                icon="info-sign"
            )
        ).add_to(mapa)

    # -------- POLIGONO ZONA GRANDE --------
    folium.Polygon(
        locations=zona_eduardo_2,
        color='blue',
        weight=5,
        fill=True,
        fill_color='blue',
        fill_opacity=0.22,
        popup='Zona Eduardo 2',
        tooltip='Zona Eduardo 2'
    ).add_to(mapa)

    # -------- POLIGONO ZONA PEQUEÑA --------
    folium.Polygon(
        locations=zona_eduardo_1,
        color='blue',
        weight=5,
        fill=True,
        fill_color='blue',
        fill_opacity=0.22,
        popup='Zona Eduardo 1',
        tooltip='Zona Eduardo 1'
    ).add_to(mapa)

    folium.Polygon(
        locations=puntos,
        color='blue',
        weight=5,
        fill=True,
        fill_color='blue',
        fill_opacity=0.22,
        popup='Zona Eduardo 1',
        tooltip='Zona Eduardo 1'
    ).add_to(mapa)

    # -------- AJUSTAR VISTA --------
    todos_los_puntos = zona_eduardo_1 + zona_eduardo_2 + puntos
    mapa.fit_bounds(todos_los_puntos)

    # -------- GUARDAR Y ABRIR --------
    archivo = "mapa.html"
    mapa.save(archivo)

    print("Mapa generado correctamente:", archivo)
    webbrowser.open("file://" + os.path.realpath(archivo))

else:
    print("No se pudieron extraer coordenadas.")
```
